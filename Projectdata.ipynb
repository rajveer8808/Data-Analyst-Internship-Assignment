
○	Total issuances and retirements per year

```python
combined_df['Issuance Date'] =pd.to_datetime(combined_df['Issuance Date'],errors='coerce')
#  exract the year
combined_df['Issuance Date']=combined_df['Issuance Date'].dt.year
combined_df['Retirement Date']=pd.to_datetime(combined_df['Retirement Date'],errors='coerce')
combined_df['Retirement Date']=combined_df['Retirement Date'].dt.year
total_issuances_per_year = combined_df.groupby('Issuance Date').size()
total_retires_per_year = combined_df.groupby('Retirement Date').size()
print(total_issuances_per_year)
print(total_retires_per_year)
```

**Top project types by issuance volume**

```python
group = combined_df.groupby('Project Type')['Quantity'].sum().reset_index()

# Now sort the values by 'Quantity'
group = group.sort_values(by='Quantity', ascending=False)
print(group)
```

**Geographical distribution of projects**

```python
geo_distribution = combined_df.groupby('Country')['Project Name'].count().reset_index()
geo_distribution.columns = ['Country', 'Number of Projects']

# Sort by number of projects
geo_distribution = geo_distribution.sort_values(by='Number of Projects', ascending=False)

# Display the geographical distribution
print(geo_distribution)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(geo_distribution['Country'], geo_distribution['Number of Projects'], color='skyblue')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Add titles and labels
plt.title('Geographical Distribution of Projects')
plt.xlabel('Country')
plt.ylabel('Number of Projects')

# Show the plot
plt.tight_layout()
plt.show()
```

**Visualize key trends using appropriate charts or graphs**

```python
# Pie chart for project types by issuance volume
project_types = combined_df.groupby('Project Type')['Quantity'].sum().reset_index()

plt.figure(figsize=(8, 8))
plt.pie(project_types['Quantity'], labels=project_types['Project Type'], autopct='%1.1f%%', startangle=160, colors=sns.color_palette('pastel'))
plt.title('Project Types by Issuance Volume')

plt.show()
```

**Total Retirements Per year**

```python
plt.figure(figsize=(10, 6))
total_retires_per_year.plot(kind='bar', color='skyblue')

# Add titles and labels
plt.title('Total Retirements per Year')
plt.xlabel('Year')
plt.ylabel('Number of Retirements')

plt.tight_layout()
plt.show()
```
